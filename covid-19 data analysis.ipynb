# Task 3.3 find the maximum infection rate for all of the countries.

```python
countries = list(corona_datasets_aggregated.index)
max_infection_rates=[]
for c in countries:
    max_infection_rates.append(corona_datasets_aggregated.loc[c].diff().max())
corona_datasets_aggregated["max_infection_rate"]=max_infection_rates
```
